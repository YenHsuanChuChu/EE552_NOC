```python
import matplotlib.pyplot as plt
import os

os.makedirs('figures', exist_ok=True)

steps = [50, 100, 150, 200, 250, 300, 350, 400]
slice_loss = [11.65, 11.39, 10.84, 10.45, 8.80, 8.72, 8.79, 8.76]
lora_loss = [2.24, 2.36, 2.37, 2.38, 2.23, 2.36, 2.37, 2.37]

plt.figure(figsize=(10, 5))
plt.plot(steps, slice_loss, label='Slice Mode', marker='o', color='red')
plt.plot(steps, lora_loss, label='LoRA Mode', marker='s', color='blue')

# Label epoch 2
plt.axvline(x=200, color='gray', linestyle='--', alpha=0.8, linewidth=2.5)
plt.annotate('Epoch 2 Starts',
             xy=(200, 3),
             xytext=(220, 3),
             color='gray',
             fontsize=11,
             fontweight='bold',
             arrowprops=dict(facecolor='gray', shrink=0.05, width=1, headwidth=6))

plt.xlabel('Training Steps')
plt.ylabel('Loss')
plt.title('MoE Fine-tuning Loss Curves: Slice vs LoRA')
plt.legend()
plt.grid(True)
plt.savefig('figures/moe_loss_curves.png')
plt.show()
```
